```python
# AirBnb listings for South New Wales, Australia
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('listings.csv')   # type(df.head())
data = pd.DataFrame(df)
dh = df.drop(['id','host_id','neighbourhood_group','latitude','longitude','last_review',
             'reviews_per_month','calculated_host_listings_count','availability_365'],axis=1)
dh
```

```python
plt.figure(figsize=(10,10))
plt.scatter(data['price'],data['neighbourhood'],color='c')    # Using data from the entire data set
plt.title('Price Given Which Neighborhood the Destination Resides')
plt.xlabel('Price')
plt.ylabel('Neighborhood')
plt.show()
# We can see the most popular destinations and the most pricey
# If I were going to visit and I only had this information to make my decision, 
# I would probably decide on a smaller town that is less expensive like Canterbury or Camden
```

```python
plt.figure(figsize=(10,10))
y=data['price']
x=data['neighbourhood']
plt.barh(x,y,color='m')
plt.show()
# The more densly populated areas cost more to stay in due to the high demand
```

```python
bins = 30
plt.hist(data['price'], bins, color='c')   
# Comparatively speaking, almost all of the locations cost less than $1500
```
